```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python

from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
veri = pd.read_csv("Mall_Customers.csv") #Veriyi yükleyelim
```

```python
X = veri.iloc[:, [3, 4]].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  #Ölçeklendirmek için
```

```python
dbi_skorlari = [] #Skorları bir listede tutmak için
```

```python
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    etiketler = kmeans.fit_predict(X_scaled)
    dbi = davies_bouldin_score(X_scaled, etiketler)
    dbi_skorlari.append(dbi)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), dbi_skorlari, marker='o', linestyle='--', color='green')
plt.title("Davies-Bouldin Skoruna Göre En İyi Küme Sayısı")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Davies-Bouldin Skoru")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as pl
```

```python
X, _ = make_blobs(n_samples=200, centers=3, random_state=42)
```

```python
kmeans1 = KMeans(n_clusters=3, random_state=42)
labels1 = kmeans1.fit_predict(X)
```

```python
X_scaled = StandardScaler().fit_transform(X)  #Ölçeklendirme yaptık
```

```python
kmeans2 = KMeans(n_clusters=3, random_state=42)
labels2 = kmeans2.fit_predict(X_scaled)
```

```python
#Grafiği çizmek için
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].scatter(X[:, 0], X[:, 1], c=labels1)
axs[0].set_title("Ölçekleme Yok")

axs[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels2)
axs[1].set_title("StandardScaler ile")

plt.tight_layout()
plt.show()

```
