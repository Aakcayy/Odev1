##  Decision Tree

Veriyi dallara ayırarak sınıflandırma veya regresyon işlemi yapabilen bir denetimli öğrenme algoritmasıdır. 
Ağaç yapısında, her düğümde bir özellik üzerinden ayrım yapılır ve yaprak düğümler son kararları verir.

Anlaşılması ve yorumlanması kolaydır ve görselleşitirilir.
Kısaca basit karar verme adımları büyük miktardaki verilere uygulanarak çok küçük kayıt gruplarına bölerek kullanılan bir yapıdır.

Eğer doğru şekilde açıklanmazsa ipin ucu kaçar hem aşırı karmaşık hal alır hem de doğruluğu azalır
Denge içinde olup düzgün açıklanmalıdır.

```python
from sklearn.datasets import load_iris  # Iris veri setini yüklüyoruz
```

```python
from sklearn.tree import DecisionTreeClassifier  # Karar ağacı modelini alıyoruz
```

```python
from sklearn.model_selection import train_test_split  # Veriyi eğitim ve test olarak ayırmak için
```

```python
iris = load_iris() #İris veri setini yükle
X = iris.data  # Özellikler 
y = iris.target  # Hedefler (Etiket)

#Etiket ile verileri birbirinden ayırdık
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
#Text size ile ne kadar eğitim için test için olsun belirlersin yoksa program kendi belirler
```

```python
# Karar ağacımızı oluşturuyoruz ve eğitiyoruz
model = DecisionTreeClassifier()  # Modeli oluşturduk
model.fit(X_train, y_train)  # Modeli eğitim verisiyle eğittik
```

```python
# Test verisiyle tahmin yapıyoruz
y_pred = model.predict(X_test)
```

```python
y_pred
```

## Random Forest

Bu algoritma, oluşturduğu bir sürü karar ağacı ile onların ortak kararına göre sonucu belirliyor. Yani çoğunluk kararını esas alıp ilerliyor.

Daha sağlam sonuç verir ve ipin ucu burada daha sağlamdır.
Birden fazla ağaç olduğundan yorumu biraz daha zordur ve bu yüzden de hesaplaması daha uzun sürer.



```python
from sklearn.ensemble import RandomForestClassifier  # Random Forest modelini alıyoruz
from sklearn.model_selection import train_test_split  # Veriyi eğitim ve test olarak ayırmak için
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model_rf = RandomForestClassifier(n_estimators=10)  # 10 ağaçla eğitiyoruz
model_rf.fit(X_train, y_train)
```

```python
y_pred_rf = model_rf.predict(X_test)
```
